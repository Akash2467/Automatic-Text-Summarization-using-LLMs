```python
from transformers import PegasusTokenizer, PegasusForConditionalGeneration

# Load the PEGASUS model and tokenizer
model_name = "google/pegasus-xsum"
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name)
```

```python
# Input text to summarize
text = """
Liverpool secured a 3-1 victory over Manchester United in a thrilling Premier League match.
Mohamed Salah scored twice, while Sadio Mané added another goal to seal the win.
Manchester United's only goal came from Marcus Rashford, but it wasn't enough to mount a comeback.
Liverpool now sits at the top of the table, three points ahead of their closest rivals.
"""
```

```python
# Tokenize the input text
inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
# Generate summary
summary_ids = model.generate(**inputs, max_length=60, min_length=20, length_penalty=2.0, num_beams=5)
generated_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
# Print the summary
print("Generated Summary:", generated_summary)
```

```python
from rouge_score import rouge_scorer

# Reference summary
reference_summary = "Liverpool beat Manchester United 3-1 with Salah's double and Mané's goal, securing the top spot in the league."

# Calculate ROUGE score
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(reference_summary, generated_summary)

# Display ROUGE scores
print("\nROUGE-1:", scores['rouge1'].fmeasure)
print("ROUGE-2:", scores['rouge2'].fmeasure)
print("ROUGE-L:", scores['rougeL'].fmeasure)
```
